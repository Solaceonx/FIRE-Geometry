```python
import plot_galaxy as gplot
import mg
import os
import time
import h5py
import numpy as np
globalparams_path = "/DFS-L/DATA/cosmo/jgmoren1/FIREBox/FB15N1024/global_sample_data/global_sample_data_snapshot_1200.hdf5"
globalparams = h5py.File(globalparams_path, 'r')
#globalparams.keys()
```
