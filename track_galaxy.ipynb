```python
import plot_galaxy as gplot
import mg
import galaxy_tools
import time
age = 0.1
snap_nums = ['020', '029', '059', '088', 102, 214, 294, 332, 412, 446, 486, 590, '041', '067', 120, 172, 277, 382, 534, 600]
```

```python
start_time = time.time()
sim_inputs = [
    'm12b_res7100',
]
for snap_num in snap_nums:
    star_center = mg.measure_m12_coords(sim_inputs, snap_num = snap_num, age_range = age)
    if star_center is not None:
        csv_file = 'm12_redshift_shapes/m12b_02/m12b_coords_' + str(snap_num) + '.csv'
        mg.save_coordinates(star_center, csv_file)
    else:
        print('RIPBOZO')
end_time = time.time() 
elapsed_time = end_time - start_time
print(f"Total time taken: {elapsed_time:.2f} seconds")
```
